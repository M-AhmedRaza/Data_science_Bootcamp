# Data Inconsitencies

1. Inconsistent format
      a. e.g date format
2. Inconsitent Naming
   1. e.g country america/usa/us/U.S.A
3. Typographical error
4. Duplication
5. Contradictory data
   1. which doest make sense, technically impossible

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={
    'date':['2019-01-01', '2020-11-02', '2022/10/13', '12-01-2019'],
    'country':['USA', 'US', 'America', 'Unied States'],
    'name':['Tom Harry', 'Tom harry', 'Tim harry', 'Tom Jerry'],
    'sales_2010':[100, 200, None, 200],
    'sales_2020':[None, 150, 300, 250]
}

# converting data to dataframe
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
#standardization of date column
df['date']=pd.to_datetime(df['date'], errors='coerce')
df['date']=df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

## How to impute NAN values via constant or variable

```python
#harmonize the country column by replacing the country names with standard names
country_mapping={
    'USA':'United States','US':'United States','America':'United States','Unied States':'United States'
}
df['country']=df['country'].replace(country_mapping)
df.head()
```

```python
# harmonize the name column by replacing the name with standard names
name_mapping={
    'Tom Harry':'Tom Harry', 'Tom harry':'Tom Harry', 'Tim harry':'Tom Harry', 'Tom Jerry':'Tom Herry'
}
df['name']=df['name'].replace(name_mapping)
df.head()
```

```python
# remove duplicates of subset
df.drop_duplicates(subset=['name'], inplace=True)
df.head()
```

```python
# resolving contradictory data
# for demonstration purpose: assume that sales_2020 is always greater than sales_2010
# removing rows where sales_2020 is less than sales_2010
# Resolving contradictory data: assume that sales_2020 is always greater than sales_2010
df = df[df['sales_2020'] > df['sales_2010']]    
df.head()

# drop using df.drop
df = df.drop(df[df['sales_2020'] < df['sales_2010']].index)
df.head()
```
