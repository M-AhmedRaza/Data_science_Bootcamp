# Example

```python
#load the dataset
df=sns.load_dataset('titanic')
df.head()
```

```python
#preprocessing
df.drop('deck',axis=1, inplace=True)

#impute missing values using imputer
# Create an instance of the SimpleImputer class
imputer = SimpleImputer()
# Impute missing values of age and fare columns using simple imputer
df[['age', 'fare']] = imputer.fit_transform(df[['age', 'fare']])


#impute missing values of embarked and embark_town
imputer = SimpleImputer(strategy='most_frequent')
df[['embarked', 'embark_town']] = imputer.fit_transform(df[['embarked', 'embark_town']])



```

```python
df.isnull().sum()
```

```python
#encode the categorical and object variable using for loop
for col in df.columns:
    if df[col].dtype=='object' or df[col].dtype=='category':
        le=LabelEncoder()
        df[col]=le.fit_transform(df[col])
```

```python
df.head()
```

```python
#drop survived and alive
X=df.drop(['survived','alive'],axis=1)
y=df['survived']
```

```python
#split and train the model
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
#model
model=DecisionTreeClassifier(criterion='entropy', max_depth=3) 
model.fit(X_train,y_train)
#prediction
y_pred=model.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#save the decision tree
from sklearn.tree import export_graphviz
export_graphviz(model, out_file='./saved_models/tree_01.dot', feature_names=X.columns, class_names=['0','1'])
```
