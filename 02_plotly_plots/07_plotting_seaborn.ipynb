```python
# data
df= sns.load_dataset('titanic')
```

```python
df.info()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df,x='age', y='fare', hue='sex')

# scatter plot is bw two nuemeric plots
```

```python
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1, font='Calibri')
plt.figure(figsize=(10,8)) #A4 size


# change the scater points colors
palette={'male':'blue', 'female':'red'}
sns.scatterplot(data=df,x='age', y='fare', hue='sex', palette=palette)

#labels
plt.xlabel('age yrs')
plt.ylabel(' Fare $')
plt.title(' Scatter Plt of Age Vs Fare')

#saving a plot
plt.savefig('scatter_plot.png',dpi=300)
plt.savefig('../scatter_plot.svg')
```

```python
# making bubbble plot uing same plot
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1, font='Calibri')
plt.figure(figsize=(10,8)) #A4 size


# change the scater points colors
palette={'male':'blue', 'female':'red'}
sns.scatterplot(data=df,x='age', y='fare', hue='sex',size='age',palette=palette)

#labels
plt.xlabel('age yrs')
plt.ylabel(' Fare $')
plt.title(' Scatter Plt of Age Vs Fare')
```

```python
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1, font='Calibri')
plt.figure(figsize=(10,8)) #A4 size


# change the scater points colors
palette={'First':'blue', 'Second':'red', 'Third':'yellow'}
sns.boxplot(data=df,x='sex', y='fare', hue='class', palette=palette)

#labels
plt.xlabel('sex')
plt.ylabel(' Fare $')
plt.title(' Scatter Plt of Age Vs Fare')
```

```python
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1, font='Calibri')
plt.figure(figsize=(10,8)) #A4 size


# change the scater points colors
palette={'First':'blue', 'Second':'red', 'Third':'yellow'}
sns.boxenplot(data=df,x='sex', y='fare', hue='class', palette=palette)

#labels
plt.xlabel('sex')
plt.ylabel(' Fare $')
plt.title(' Scatter Plt of Age Vs Fare')
```

```python
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1, font='Calibri')
plt.figure(figsize=(4,8)) #A4 size


# change the scater points colors
palette={'First':'blue', 'Second':'red', 'Third':'yellow'}
sns.violinplot(data=df,x='sex', y='fare', hue='class', palette=palette)

#labels
plt.xlabel('sex')
plt.ylabel(' Fare $')
plt.title(' Scatter Plt of Age Vs Fare')
```
