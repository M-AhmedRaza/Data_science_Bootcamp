```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=sns.load_dataset('titanic')
```

```python
#scatter plot
fig=px.scatter(df,x='age', y='fare',title= 'Titanic Plot Age vs Fare')
fig.show()
```

```python
#interactive upon click
fig=px.scatter(df,x='age', y='fare',color='sex',title= 'Titanic Plot Age vs Fare')
fig.show()
```

```python
df.dropna(inplace=True)
#making bubble plot but it requires no null values
fig=px.scatter(df,x='age', y='fare',color='sex',size='fare',title= 'Titanic Plot Age vs Fare')
fig.show()
fig.write_image('/scatter_plot2.png', format='png', scale=3)
```
